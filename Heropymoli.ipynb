```python
 # Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
#file_to_load = "/Users/lisahayman/Desktop/homework/myfolder/Heropymoli.csv"
file_to_load = "../myfolder/Heropymoli.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)

df.head()
```

```python
#   Player Count/Number of Player 

Player_Count = df["SN"].nunique()
Player_df= pd.DataFrame({
    "Player Count": [Player_Count]})

Player_df
```

```python
#  Purchasing Analysis (Total)

#Number of Unique Items
Unique_Names = df["Item Name"].nunique()
#Average Purchase Price
Average_Price = round(df["Price"].mean(),2)

#Total Number of Purchases
Number_Purchases = df["Purchase ID"].count()

#Total Revenue
Total_Revenue = df["Price"].sum()

Purchase_df= pd.DataFrame({
    "Unique Names": [Unique_Names], 
    "Average Price": [Average_Price], 
    "Number of Purchases": [Number_Purchases],
    "Total Revenue": [Total_Revenue]})

Purchase_df
```

```python
#Gender Demographics
dataframe_gender=pd.DataFrame(df["Gender"].value_counts()) 
dataframe_gender


#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

gender_count = dataframe_gender["Gender"].sum()
dataframe_gender["Gender Percentage"] = round((dataframe_gender["Gender"]/gender_count)*100,2)
dataframe_gender

```

```python
#Purchasing Analysis (Gender)

#The below each broken by gender

#Purchase Count
purchase_count = df.groupby('Gender').count()['Purchase ID']

#Average Purchase Price
purchase_price = df.groupby('Gender').mean()['Price']

#Total Purchase Value
total_value = df.groupby('Gender').sum()['Price']

#Average Purchase Total per Person by Gender

Avg_purchase_PP = total_value/purchase_count 

Gender_Analysis = pd.DataFrame({
        "Purchase Count": purchase_count, 
        "Purchase Price": purchase_price,
        "Total Value": total_value,
        "Average Purchase Total per Person by Gender": Avg_purchase_PP})

Gender_Analysis



```

```python
#Age Demographics

#age_bins = [0, 9.99, 13.99, 18.99, 23.99, 28.99, 33.99, 38.99, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins
df["Age Ranges"] = pd.cut(df["Age"], age_bins, labels=group_names)
df.head()

#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
```

```python
 

# Calculate the numbers and percentages by age group

age_group_df = pd.DataFrame(df["Age Ranges"].value_counts()) 

age_range = age_group_df["Age Ranges"].sum()

age_group_df["Pecentage of Players"] = round((age_group_df["Age Ranges"]/age_range)*100,2)
age_group_df.sort_index(inplace=True)

age_group_df
```

```python
#Average Purchase Price
Avg_PP = round(df.groupby('Age Ranges').mean()['Price'], 2)
Total_PV = df.groupby('Age Ranges').sum()['Price']
Purchase_Count = round(df.groupby('Age Ranges').count()['Purchase ID'], 2)

Avg_Total_pp = round(Total_PV/Purchase_Count,2)

Age_Analysis = pd.DataFrame({
        "Average Purchase Price": Avg_PP, 
        "Total Purchase Value": Total_PV,
        "Purchase Count": Purchase_Count,
        "Avg Total Purchase per Person": Avg_Total_pp})

Age_Analysis
```

```python
#Total Purchase Value
total_value = df.groupby('Gender').sum()['Price']

#Average Purchase Total per Person by Gender

Avg_purchase_PP = total_value/purchase_count 

Gender_Analysis = pd.DataFrame({
        "Purchase Count": purchase_count, 
        "Purchase Price": purchase_price,
        "Total Value": total_value,
        "Average Purchase Total per Person by Gender": Avg_purchase_PP})

Gender_Analysis


#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Average Purchase Total per Person by Age Group
```
